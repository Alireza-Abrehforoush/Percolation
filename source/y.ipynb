```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data2.csv')
```

```python
grouped = df.groupby(['L/2', 'density', 'probability_of_spread', 'south_wind_speed', 'west_wind_speed', 'big-jumps'])

# Calculate the average of 'percentage-burned' within each group
df_avg = grouped['percentage-burned'].mean().reset_index()
```

```python
list_of_big_jump = [False, True]
list_of_south_wind_speed = [-20, -10, 0, 10, 20]
list_of_west_wind_speed = [-20, -10, 0, 10, 20]
list_of_probability_of_spread = [20, 40, 60, 80, 100]
list_of_density = [10, 20, 30, 40, 50, 60, 70, 80, 90]

for i in list_of_big_jump:
    # Filter conditions
    filtered_result = df_avg[
        (df_avg['L/2'] == 450) &
        (df_avg['probability_of_spread'] == 100) &
        (df_avg['south_wind_speed'] == 0) &
        (df_avg['west_wind_speed'] == 0) &
        (df_avg['big-jumps'] == i)
    ]
    plt.figure(figsize=(10, 6))  # Set the figure size
    # Line plot
    plt.plot(filtered_result['density'], filtered_result['percentage-burned'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
    # Add text annotation
    text_to_display = f"L/2: {450}\nProb. of Spread: {100}\nSouth Wind: {0}\nWest Wind: {0}\nBig Jumps: {i}"
    plt.text(0.02, 0.8, text_to_display, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))
    # Labels and title
    plt.xlabel('Density')
    plt.ylabel('Percentage Burned')
    plt.title('Percentage Burned vs. Density for Filtered Conditions')
    # Save the plot as an image file (e.g., PNG)
    plt.grid(True)
    file_name = f"BIG_JUMPS_________L_2[{450}]-Prob. of Spread[{0}]-South Wind[{0}]-West Wind[{0}]-Big Jumps[{i}]"
    plt.savefig("./plots1/" + file_name + ".png")




for i in list_of_south_wind_speed:
    # Filter conditions
    filtered_result = df_avg[
        (df_avg['L/2'] == 450) &
        (df_avg['probability_of_spread'] == 100) &
        (df_avg['south_wind_speed'] == i) &
        (df_avg['west_wind_speed'] == 0) &
        (df_avg['big-jumps'] == False)
    ]
    plt.figure(figsize=(10, 6))  # Set the figure size
    # Line plot
    plt.plot(filtered_result['density'], filtered_result['percentage-burned'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
    # Add text annotation
    text_to_display = f"L/2: {450}\nProb. of Spread: {100}\nSouth Wind: {i}\nWest Wind: {0}\nBig Jumps: {False}"
    plt.text(0.02, 0.8, text_to_display, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))
    # Labels and title
    plt.xlabel('Density')
    plt.ylabel('Percentage Burned')
    plt.title('Percentage Burned vs. Density for Filtered Conditions')
    # Save the plot as an image file (e.g., PNG)
    plt.grid(True)
    file_name = f"SOUTH_WIND_SPEED_________L_2[{450}]-Prob. of Spread[{0}]-South Wind[{i}]-West Wind[{0}]-Big Jumps[{False}]"
    plt.savefig("./plots1/" + file_name + ".png")



for i in list_of_west_wind_speed:
    # Filter conditions
    filtered_result = df_avg[
        (df_avg['L/2'] == 450) &
        (df_avg['probability_of_spread'] == 100) &
        (df_avg['south_wind_speed'] == 0) &
        (df_avg['west_wind_speed'] == i) &
        (df_avg['big-jumps'] == False)
    ]
    plt.figure(figsize=(10, 6))  # Set the figure size
    # Line plot
    plt.plot(filtered_result['density'], filtered_result['percentage-burned'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
    # Add text annotation
    text_to_display = f"L/2: {450}\nProb. of Spread: {100}\nSouth Wind: {0}\nWest Wind: {i}\nBig Jumps: {False}"
    plt.text(0.02, 0.8, text_to_display, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))
    # Labels and title
    plt.xlabel('Density')
    plt.ylabel('Percentage Burned')
    plt.title('Percentage Burned vs. Density for Filtered Conditions')
    # Save the plot as an image file (e.g., PNG)
    plt.grid(True)
    file_name = f"WEST_WIND_SPEED_________L_2[{450}]-Prob. of Spread[{0}]-South Wind[{0}]-West Wind[{i}]-Big Jumps[{False}]"
    plt.savefig("./plots1/" + file_name + ".png")



for i in list_of_probability_of_spread:
    # Filter conditions
    filtered_result = df_avg[
        (df_avg['L/2'] == 450) &
        (df_avg['probability_of_spread'] == i) &
        (df_avg['south_wind_speed'] == 0) &
        (df_avg['west_wind_speed'] == 0) &
        (df_avg['big-jumps'] == False)
    ]
    plt.figure(figsize=(10, 6))  # Set the figure size
    # Line plot
    plt.plot(filtered_result['density'], filtered_result['percentage-burned'], marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
    # Add text annotation
    text_to_display = f"L/2: {450}\nProb. of Spread: {i}\nSouth Wind: {0}\nWest Wind: {0}\nBig Jumps: {False}"
    plt.text(0.02, 0.8, text_to_display, transform=plt.gca().transAxes, bbox=dict(facecolor='white', alpha=0.5))
    # Labels and title
    plt.xlabel('Density')
    plt.ylabel('Percentage Burned')
    plt.title('Percentage Burned vs. Density for Filtered Conditions')
    # Save the plot as an image file (e.g., PNG)
    plt.grid(True)
    file_name = f"PROBABILITY_OF_SPREAD_________L_2[{450}]-Prob. of Spread[{i}]-South Wind[{0}]-West Wind[{0}]-Big Jumps[{False}]"
    plt.savefig("./plots1/" + file_name + ".png")

```
